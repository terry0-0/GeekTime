```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)

plt.scatter(x, y, marker="x")
plt.show()


df = pd.DataFrame({'x':x, 'y':y})
sns.jointplot(x="x", y="y", data=df, kind='scatter')
plt.show()
```
