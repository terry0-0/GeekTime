```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5,3,6,20,17,16,19,30,32,35]

plt.plot(x, y)
plt.show()

df = pd.DataFrame({'x':x, 'y':y})
sns.lineplot(x="x", y="y", data=df)
plt.show()
```
